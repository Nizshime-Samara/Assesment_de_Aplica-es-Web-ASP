7-Escreva um programa em Python que realiza operações de inclusão e remoção 
em listas. Seu programa deve perguntar ao usuário qual operação deseja fazer: (código)


*  a) Mostrar lista;
*  b)Incluir elemento;
*  c)Remover elemento;
*  d)Apagar todos os elementos da lista.

Se a opção for a alternativa (a), seu programa deve apenas mostrar o conteúdo da lista. Se a opção for a alternativa (b), seu programa deve pedir o valor do elemento a ser incluído. Se a opção for a alternativa (c), seu programa deve pedir o valor do elemento a ser removido. E se a opção for a alternativa (d), deve-se apenas exibir se a operação foi concluída.




```

lista = ["Abacate", "Abacaxi", "Acerola", "Amora", "Banana", "Cacau", "Caja", "Caqui", "Carambola", "Cereja"]

menu_options = '''
a) Mostrar lista
b) Incluir elemento
c) Remover elemento
d) Apagar todos os elementos da lista
z) Encerrar execução
'''
done = False

def menu():
    escolha = input(menu_options + "\nEscolha uma opção do menu acima: ")
    if escolha == "a":
        print(*lista, sep=", ")
    elif escolha == "b":
        incluir_elemento = input("Qual elemento devo incluir: ")
        lista.append(incluir_elemento)
    elif escolha == "c":
        remove_elemento = input("Qual elemento devo remover: ")
        lista.remove(remove_elemento)
    elif escolha == "d":
        try:
            lista.clear()
        except:
            print("A lista não pode ser limpa!")
        else:
            print("A lista foi limpa com sucesso!")
    elif escolha == "z":
        return True
    else:
        print("Opção inválida!")
    return False

while done == False:
    done = menu()
```

9-Usando a biblioteca Pygame, escreva um programa que implemente uma interface para o gerenciamento de informações de hardware: memória, disco(s)(se houver mais de um), cpu: 



```
import pygame
import psutil
import cpuinfo

pygame.font.init()

font = pygame.font.Font(None, 32)
azul = (0, 0, 255)
preto = (0, 0, 0)
vermelho = (255, 0, 0)
amarelo = (255, 255, 0)
verde = (0, 153, 0)
branco = (255, 255, 255)
larg = 0

largura_tela = 800
altura_tela = 600
tela = pygame.display.set_mode((largura_tela, altura_tela))
pygame.display.set_caption("Exemplo Surface")
pygame.display.init()

def mostra_infos():
    disco = psutil.disk_usage('.')
    larg = largura_tela - 2*20
    tela.fill(preto)
    s1 = pygame.surface.Surface((largura_tela, altura_tela/3))
    pygame.draw.rect(s1, azul, (20, 50, largura_tela-2*20, 70))
    tela.blit(s1, (0, 0))
    larg = larg*disco.percent/100
    pygame.draw.rect(tela, vermelho, (20, 50, larg, 70))
    total = round(disco.total/(1024*1024*1024), 2)
    texto_barra = "Uso de Disco: (Total: " + str(total) + "GB):"
    text = font.render(texto_barra, 1, branco)
    tela.blit(text, (20, 10))

    capacidade = psutil.cpu_percent(interval=0)
    s2 = pygame.surface.Surface((largura_tela, altura_tela/3))
    pygame.draw.rect(s2, azul, (20, 50, largura_tela-2*20, 70))
    tela.blit(s2, (0, altura_tela/3))
    larg = larg*capacidade/100
    pygame.draw.rect(tela, amarelo, (20, 250, larg, 70))
    text = font.render("Uso de CPU:", 1, branco)
    tela.blit(text, (20, 200))

    mem = psutil.virtual_memory()
    s3 = pygame.surface.Surface((largura_tela, altura_tela/3))
    pygame.draw.rect(s3, azul, (20, 50, largura_tela-2*20, 70))
    tela.blit(s3, (0, 2*altura_tela/3))
    larg = larg*mem.percent/100
    pygame.draw.rect(tela, verde, (20, 450, larg, 70))
    percentagem = mem.percent
    total = round(mem.total/(1024*1024*1024),2)
    #texto_barra = "Uso de Memória (Total: " + str(total) + "GB):"
    texto_barra1 = ('{}%(Total:{}GB)'.format(percentagem,total))
    text = font.render(texto_barra1, 1, branco)
    tela.blit(text, (20, 400))
    

clock = pygame.time.Clock()
terminou = False
# inicia a variavel que quando for igual a 60 a informação será mostrada
cont = 30
while not terminou:
    # Atualiza o desenho na tela
    pygame.display.update()
    # monitora os eventos na tela
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            terminou = True
        # Fazer a atualização a cada segundo:
    if cont == 30:
        cont = 0
        mostra_infos()
        

# 60 frames por segundo
    clock.tick(30)
    cont = cont + 1
# Finaliza a janela
pygame.display.quit()
```

(Desafio) Usando a biblioteca Pygame, escreva um programa que implemente o jogo da memória para um jogador.

O jogo da memória é um clássico jogo formado por peças que apresentam uma figura ou cor em um dos lados. Cada figura/cor se repete em duas peças diferentes. Para começar o jogo, as peças são postas com as figuras/cores voltadas para baixo, para que não possam ser vistas. Cada jogada do jogador consiste em virar duas peças, clicando nelas. Caso as figuras/cores sejam iguais, as peças permanecem viradas. Se forem peças diferentes, estas devem ser viradas novamente. O jogo termina quando todas as peças forem viradas. A pontuação do jogador corresponde a todas as suas jogadas. Quanto menor o número de jogadas, melhor a sua pontuação.

As peças podem ser quadrados posicionadas em forma de matriz. Por exemplo, se forem 16 peças, elas podem ser posicionadas na resolução 4x4 ou 2x8 ou como preferir.
